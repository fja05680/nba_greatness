Other imports

```python
import pandas as pd
import numpy as np
import re
import datetime
```

open the cap file and read lines into a list

```python
start = True
df = pd.DataFrame(columns=['mxc_restart_dt', 'uboot_dt'])

with open('4.11.2018minicom.cap', 'r') as f:
    for line in f:
        # [2018-04-06 11:34:50] mxc_restart: Using gpio to reboot
        m = re.match(r'\[(\d{4}\-\d{2}\-\d{2} \d{2}:\d{2}:\d{2})\] mxc_restart: Using gpio to reboot', line)
        if m:
            mxc_restart_dt=datetime.datetime.strptime(m.group(1), '%Y-%m-%d %H:%M:%S')
            df2 = pd.DataFrame([mxc_restart_dt], columns=['mxc_restart_dt'])
            df = df.append(df2, ignore_index=True)
            start = False
        
        if not start:
            # [2018-04-06 11:34:50] U-Boot 2015.10+fslc+g1b6aee7 (Mar 08 2018 - 18:48:18 -0500)
            m = re.match(r'\[(\d{4}\-\d{2}\-\d{2} \d{2}:\d{2}:\d{2})\] U-Boot 2015.+', line)
            if m:
                uboot_dt=datetime.datetime.strptime(m.group(1), '%Y-%m-%d %H:%M:%S')
                df.loc[df['mxc_restart_dt'] == mxc_restart_dt, 'uboot_dt'] = uboot_dt
                start = True             
                     
                     
                     
                     
                     
```

```python
df
```

Check for NaN columns

```python
pd.isnull(df).sum()
```

Fill NaN Columns

```python
if pd.isnull(df['uboot_dt']).sum() > 0:
    df['uboot_dt'].fillna(df['mxc_restart_dt'])
```

convert column to datetime

```python
df['uboot_dt'] = pd.to_datetime(df['uboot_dt'])
```

calculate diff between uboot_dt and mxc_restart_dt

```python
df.dtypes
df['diff_uboot_mxc'] = df['uboot_dt'] - df['mxc_restart_dt']
df['diff_uboot_mxc'] = df['diff_uboot_mxc'].dt.total_seconds()
```

calc diff for mxc_restart_dt

```python
df.dtypes
df['diff_mxc2_mxc1'] = df['mxc_restart_dt'].diff()
df['diff_mxc2_mxc1'] = df['diff_mxc2_mxc1'].dt.total_seconds()
```
