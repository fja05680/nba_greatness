```python
from __future__ import print_function
from __future__ import unicode_literals
from __future__ import division
from __future__ import absolute_import

import numpy as np
import pandas as pd
import os
import re
import shutil
import wikipedia as wp

#html = wp.page('charles barkley').html()
#html = wp.page('magic johnson').html()
#html = wp.page('julius erving').html()
html = wp.page('steve kerr').html()
#html = wp.page('bill russell').html()

df = pd.read_html(html, match='Career highlights and awards', attrs={'class':'infobox'}, index_col=0)
```

```python
df = df[0]
df
```

```python
v = df.loc[['Position']]
s = v[1]
s
```

```python
positions = ['CENTER', 'FORWARD', 'GUARD']
for pos in s:
    if any(position in pos.upper() for position in positions):
        print(pos)
```

```python
v = df.loc[['Playing career']]
s = v[1]
print(s[0])
```

```python
v = df[df.index.str.contains('NBA champion')==True]
str = v.index[0]
str
```

```python
m1 = re.match(r'.*player:?\s+(\d+).+NBA champion.*', str)
m2 = re.match(r'.*coach:?\s+(\d+).+NBA champion.*', str)
m3 = re.match(r'.*(\d+).+NBA champion.*', str)
m4 = re.match(r'.*NBA champion.*', str)
if m1:
    num = m1.group(1)
    print('player', num)
if m2:
    num = m2.group(1)
    print('coach', num)
elif m3:
    num = m3.group(1)
    print('player', num)
elif m4:
    print('player once', 1)
```
